## 📌 Bloco 2 — O que é um CSV?

**CSV** = Comma-Separated Values (Valores Separados por Vírgula)

É um arquivo de texto simples onde:
- Cada **linha** é uma linha da tabela
- Cada **coluna** é separada por vírgula

Exemplo:
```
date,product,category,quantity,price
2024-01-15,Laptop,Electronics,2,999.99
2024-01-16,Mouse,Electronics,5,29.99
```

💡 Toda planilha Excel pode ser salva como CSV. Sistemas de vendas exportam CSV. É o formato mais comum de dados no mundo real.

## 📌 Bloco 3 — Lendo o CSV com pandas

**pandas** é a biblioteca do Python feita para trabalhar com dados em tabela.
Como se fosse o Excel do Python.

`df` vem de **DataFrame** — o nome que o pandas dá para uma tabela.

````python
# Importando a biblioteca pandas
import pandas as pd

url = 'https://raw.githubusercontent.com/mgdossantos/automated_sales_report/main/data/sales.csv'
df = pd.read_csv(url)
print("Dados carregados! ✅")
````

````python
# Visualizando a tabela completa
print(df)
````

---
## 📌 Bloco 4 — Explorando os dados

Sempre que você não sabe o que tem dentro de um DataFrame, use `.head()`.
É como dar uma espiada nos dados antes de trabalhar com eles.

````python
# Quantas linhas e colunas?
print("Tamanho do dataset:")
print(df.shape)
print(f"→ {df.shape[0]} linhas e {df.shape[1]} colunas")
````

````python
# Quais são as colunas?
print("Colunas disponíveis:")
print(df.columns.tolist())
````

````python
# Primeiras 5 linhas
print("Primeiras 5 linhas:")
df.head()
````

````python
# Tipo de dado de cada coluna
print("Tipos de dados:")
print(df.dtypes)
print()
print("💡 A coluna 'date' aparece como 'object' (texto), não como data.")
print("   Isso é muito comum em dados reais — vamos falar sobre isso em outra live!")
````

## 📌 Bloco 5 — Extraindo insights reais

Com as colunas `date`, `product`, `category`, `quantity` e `price` já conseguimos responder perguntas reais de negócio.

**Vamos criar a coluna de receita primeiro:**

````python
# Calculando a receita por transação
# receita = quantidade × preço
df['revenue'] = df['quantity'] * df['price']

print("Nova coluna 'revenue' criada! ✅")
df.head()
````

````python
# Receita total
receita_total = df['revenue'].sum()
print(f"💰 Receita total: R$ {receita_total:,.2f}")
````

````python
# Produto mais vendido por quantidade
print("📦 Produtos mais vendidos (por quantidade):")
print()
print(df.groupby('product')['quantity'].sum().sort_values(ascending=False))
````

````python
# Ticket médio por transação
ticket_medio = df['revenue'].mean()
print(f"🎯 Ticket médio por transação: R$ {ticket_medio:.2f}")
````

````python
# Resumo geral dos dados numéricos
print("📊 Resumo estatístico:")
df[['quantity', 'price', 'revenue']].describe().round(2)
````

````python
# Calculando a receita total por produto
receita_por_produto = df.groupby('product')['revenue'].sum().sort_values(ascending=False)

print("💰 Receita por Produto (do maior para o menor):")
print(receita_por_produto)
````

## 📌 Bloco 6 — Gerando gráficos

Um gráfico de barras de receita por produto


````python
import plotly.express as px

# Convert the Series to a DataFrame so 'product' and 'revenue' become explicit columns
receita_por_produto_df = receita_por_produto.reset_index()
print(receita_por_produto_df)
````

````python
fig = px.bar(
    receita_por_produto_df,
    x = "revenue",
    y="product", # Changed to 'product' as this is the correct column name
    title= "Receita por Produto", # Updated title to be relevant
    color ="product" # Changed to 'product' as this is the correct column name
)
fig.show()
````
