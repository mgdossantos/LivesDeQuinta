## 📌 Bloco 1 — Instalando e importando tudo

Primeiro vamos importar as bibliotecas que vamos usar.

`matplotlib` → faz gráficos  
`ipywidgets` → cria campos interativos no Colab  
`IPython.display` → mostra os widgets na tela

```
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

print('✅ Bibliotecas importadas com sucesso!')
```

---
## 📌 Bloco 2 — Variáveis: o que são e como usamos aqui

Uma variável é uma caixinha com nome que guarda um valor.

Exemplos:

```
# Variáveis de texto
nome = "Marcela"
objetivo = "viagem para o Japão"

# Variáveis numéricas
meta = 5000.0       # quanto quero economizar (R$)
meses = 12          # em quantos meses
renda = 3000.0      # renda mensal (R$)

# Calculando: quanto preciso guardar por mês?
por_mes = meta / meses

print(f'Nome: {nome}')
print(f'Objetivo: {objetivo}')
print(f'Meta: R$ {meta:.2f}')
print(f'Prazo: {meses} meses')
print(f'Guardar por mês: R$ {por_mes:.2f}')
```

---
## 📌 Bloco 3 — if/else: tomando decisões

O `if/else` é como uma bifurcação na estrada.  
O Python escolhe um caminho dependendo da condição.

Vamos usar para saber se a meta é viável:

```
# Quanto a pessoa pode guardar? (sugerimos 30% da renda)
capacidade = renda * 0.30

print(f'Você pode guardar até: R$ {capacidade:.2f}/mês (30% da renda)')
print(f'Você precisa guardar:  R$ {por_mes:.2f}/mês')
print()

if por_mes <= capacidade:
    print('✅ Meta VIÁVEL! Você consegue!')
elif por_mes <= renda * 0.50:
    print('⚠️  Meta DESAFIADORA. Vai precisar de esforço.')
else:
    print('❌ Meta DIFÍCIL com esse prazo. Considere aumentar o prazo.')
```

---
## 📌 Bloco 4 — Gráfico do progresso

Agora vamos visualizar o crescimento da poupança mês a mês:

```
# Criando os dados do gráfico
meses_lista = list(range(0, meses + 1))
poupanca = [por_mes * m for m in meses_lista]

# Desenhando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(meses_lista, poupanca, color='#1e3a8a', linewidth=2.5, marker='o', markersize=5)
plt.axhline(y=meta, color='#e74c3c', linestyle='--', linewidth=2, label=f'Meta: R$ {meta:.2f}')
plt.fill_between(meses_lista, poupanca, alpha=0.15, color='#1e3a8a')

plt.title(f'Jornada para: {objetivo}', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Poupança acumulada (R$)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f'\n🎯 Em {meses} meses você terá: R$ {poupanca[-1]:.2f}')
```

## 🚀 Bloco 5 — BÔNUS: versão interativa com widgets

Agora vamos transformar em um mini app onde você muda os valores com sliders!

```
# Widgets de entrada
w_objetivo = widgets.Text(
    value='viagem para o Japão',
    description='Objetivo:',
    layout=widgets.Layout(width='400px')
)

w_meta = widgets.FloatSlider(
    value=5000, min=500, max=50000, step=500,
    description='Meta (R$):',
    readout_format='.0f',
    layout=widgets.Layout(width='400px')
)

w_meses = widgets.IntSlider(
    value=12, min=1, max=36, step=1,
    description='Prazo (meses):',
    layout=widgets.Layout(width='400px')
)

w_renda = widgets.FloatSlider(
    value=3000, min=1000, max=20000, step=500,
    description='Renda (R$):',
    readout_format='.0f',
    layout=widgets.Layout(width='400px')
)

btn = widgets.Button(
    description='Calcular minha meta! 🚀',
    button_style='primary',
    layout=widgets.Layout(width='220px', height='40px')
)

output = widgets.Output()

def calcular(b):
    with output:
        clear_output(wait=True)

        objetivo = w_objetivo.value
        meta     = w_meta.value
        meses    = w_meses.value
        renda    = w_renda.value

        por_mes    = meta / meses
        capacidade = renda * 0.30

        # Resultado textual
        print(f'💰 Meta: R$ {meta:.2f} em {meses} meses')
        print(f'📅 Guardar por mês: R$ {por_mes:.2f}')
        print(f'📊 Sua capacidade (30% da renda): R$ {capacidade:.2f}')
        print()

        if por_mes <= capacidade:
            print('✅ Meta VIÁVEL! Você consegue!')
            cor_status = '#1e3a8a'
        elif por_mes <= renda * 0.50:
            print('⚠️  Meta DESAFIADORA. Vai precisar de esforço.')
            cor_status = '#f0a500'
        else:
            print('❌ Meta DIFÍCIL com esse prazo. Considere aumentar o prazo.')
            cor_status = '#e74c3c'

        # Gráfico
        meses_lista = list(range(0, meses + 1))
        poupanca    = [por_mes * m for m in meses_lista]

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

        # Gráfico de linha — evolução
        ax1.plot(meses_lista, poupanca, color='#1e3a8a', linewidth=2.5, marker='o', markersize=4)
        ax1.axhline(y=meta, color='#e74c3c', linestyle='--', linewidth=2, label=f'Meta: R$ {meta:.0f}')
        ax1.fill_between(meses_lista, poupanca, alpha=0.15, color='#1e3a8a')
        ax1.set_title(f'Jornada: {objetivo}', fontsize=13, fontweight='bold')
        ax1.set_xlabel('Meses')
        ax1.set_ylabel('Poupança acumulada (R$)')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Gráfico de barras — distribuição da renda
        categorias  = ['Guardar\npor mês', 'Resto\nda renda']
        valores     = [por_mes, max(0, renda - por_mes)]
        cores       = [cor_status, '#e8e0cc']
        ax2.bar(categorias, valores, color=cores, edgecolor='white', linewidth=1.5)
        ax2.set_title('Distribuição da renda', fontsize=13, fontweight='bold')
        ax2.set_ylabel('Valor (R$)')
        ax2.grid(True, alpha=0.3, axis='y')
        for i, v in enumerate(valores):
            ax2.text(i, v + 30, f'R$ {v:.0f}', ha='center', fontweight='bold', fontsize=11)

        plt.tight_layout()
        plt.show()

btn.on_click(calcular)

display(
    widgets.VBox([
        widgets.HTML('<h3 style="color:#1e3a8a">🎯 Minha Calculadora de Meta</h3>'),
        w_objetivo, w_meta, w_meses, w_renda,
        btn, output
    ])
)
```
